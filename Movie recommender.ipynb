```python
import pandas as pd
import numpy  as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies.info()
```

```python
movies=movies[['movie_id','genres','keywords','title','overview','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convertcast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convertcast)
```

```python
movies.head()
```

```python
def getdirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(getdirector)
movies.head()
```
